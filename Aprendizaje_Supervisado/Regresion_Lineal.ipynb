# Regresion Lineal
La regresión lineal es una técnica estadística que se utiliza para encontrar la relación entre variables y predecir valores desconocidos.

Las técnicas de regresión lineal permiten crear un modelo lineal. Este modelo describe la relación entre una variable dependiente $y$ (también conocida como la respuesta) como una función de una o varias variables independientes $X_i$ (denominadas predictores). La ecuación general correspondiente a un modelo de regresión lineal es:
$$y = \beta_0 + \sum_{i=1}^{k} \beta_k X_k + \epsilon_i$$
donde $\beta$ representa las estimaciones de parámetros lineales que se deben calcular y
 $\epsilon$ representa los términos de error.

**¿Para qué se utiliza?**
* Para comprender y predecir el comportamiento de sistemas complejos. 
* Para analizar datos experimentales, financieros y biológicos. 
* Para predecir el precio de la vivienda en función de su tamaño. 
* Para predecir el peso de una persona en función de su altura. 
* Para establecer la relación entre la dosis de los medicamentos y la presión arterial de los pacientes. 
* Para crear un modelo para estimar los costes de las reclamaciones de seguros


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.linear_model import LinearRegression  # Libreria para generar regresiones lineales
from sklearn.datasets import make_regression       # Libreria para generar un dataset
```

```python
# Generamos un dataset con make_regression
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html#sklearn.datasets.make_regression
X, y = make_regression(
    n_samples=100,
    n_features=1,
    noise=20,
    random_state=42
)
print("Dimensiones de X: ", X.shape)
print("Dimensiones de y: ", y.shape)
```

```python
# Mostramos un grafico de los puntos generados
plt.scatter(X,y)
plt.title("Datos de prueba generados");
```

```python
# Creamos nuestro modelo de regresion lineal y lo entrenamos
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# Graficamos nuestra regresion lineal
plt.plot(X, modelo.predict(X), color="red")  # Graficamos la linea de regresion lineal
plt.scatter(X, y)                            # Graficamos los puntos de prueba
plt.title("Modelo de Regresion Lineal");
```

```python
# Imprimimos algunos parametros de la recta
print("Parametros de la recta")
print("Ec. Recta: y = mx + b")
print("Intercepcion (b): ", modelo.intercept_)
print("Pendiente (m): ", modelo.coef_)
```

## Ejemplo de Regresion Lineal Simple
Se necesita tener un modelo para predecir la cantidad de bultos de cemento necesarios para construir tantos metros cuadrados de muro

```python
# Declaración del DataFrame con datos específicos manualmente
df = pd.DataFrame({
    'metros_cuadrados': [5.7, 5.0, 5.3, 7.5, 7.0, 7.8, 10.0, 10.7, 9.6, 12.3, 12.0, 12.9, 17.0, 15.5, 15.0, 22.3, 19.0, 20.0, 27.0, 25.5, 24.0, 30.4, 28.9, 33.2],
    'bultos_cemento': [3, 2, 3, 4, 3, 4, 5, 6, 5, 6, 7, 6, 8, 7, 7, 10, 9, 10, 13, 12, 12, 15, 14, 16]
})

# Visualizacion de los datos
plt.scatter(df["metros_cuadrados"], df["bultos_cemento"])
plt.title("Relación Area de muro (m²) y Bultos de cemento")
plt.xlabel("Área del Muro (m²)")
plt.ylabel("Bultos de cemento");
```

```python
# Separamos los datos
X = df[["metros_cuadrados"]]
y = df["bultos_cemento"]
print("Dimension de X: ", X.shape)
print("Dimension de y: ", y.shape)
```

```python
# Creamos y entrenamos nuestro modelo de regresion lineal simple
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# Graficamos nuestra regresion lineal
plt.plot(X, modelo.predict(X), color="red")  # Graficamos la linea de regresion lineal
plt.scatter(X, y)                            # Graficamos los datos originales
plt.title("Regresion Lineal [Area de muro (m²) y Bultos de cemento]")
plt.xlabel("Área del Muro (m²)")
plt.ylabel("Bultos de cemento")
plt.grid(True);
```

```python
# Predecimos un la cantidad de bultos necesarios para un muro en especifico
area = 20.6
print(f"Bultos necesarios para un muro de {area}m² = {modelo.predict([[area]])}")
```
